```python
import statistics 
df = num_data.values.tolist()
for i in range(len(data[0])):  # Iterate over columns
    column_data = [row[i] for row in df]
    print(f"Column {i + 1} :")
    print(f"Mean: {statistics.mean(column_data)}")
    print(f"Median: {statistics.median(column_data)}")
    print(f"Mode: {statistics.mode(column_data)}")
```

```python
data = pd.read_csv("/Users/aarushinishantsingh/Downloads/iris/iris.csv")
print("Column names in the dataset:", data.columns)
data.loc[0, 'sepal.length'] = None
data.loc[10, 'sepal.width'] = None
print(data.isnull().sum())
```
